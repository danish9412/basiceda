```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read csv file into a pandas dataframe
df = pd.read_csv("./sample.csv")
```

```python
#Removing unnamed columns from the file
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head(3)
```

```python
# Getting percentage of missing values in each columns
missing_values_percent = df.isnull().sum()/len(df)
```

```python
dup_id = df.duplicated(subset='uniq_id', keep='first').sum()
print("Number of duplicate unique IDs :",dup_id)
```

```python
# Plotting percentage graph
percent_graph = missing_values_percent.plot(kind='barh',title='Percentage of missing data')
```

```python
# Getting percentage of missing values in each columns
missing_values = df.isnull().sum()
# Plotting percentage graph
value_graph = missing_values.plot(kind='barh',title='Number of missing data')
```
